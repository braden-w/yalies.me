```python
# Convert to dictionary
facebook_filtered_just_cities = facebook_filtered_just_cities.to_dict('records')
len(facebook_filtered_just_cities)
```

```python
# For each filtered place, get the place ID, description, longitude, and latitude from the Google Places API using the address field
import requests
import json

facebook_filtered_with_place_id = []
failed_api_calls = []
# For f in rows 1 of facebook_filtered_1_new_lines
for f in facebook_filtered_just_cities[100:]:
	try:
		address = f['address']
		url = f'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={address}&inputtype=textquery&fields=place_id,name,geometry&key={google_places_key}'
		response = requests.get(url)
		data = json.loads(response.text)
		first_result = data['candidates'][0]
		place_id = first_result['place_id']
		lat = first_result['geometry']['location']['lat']
		lng = first_result['geometry']['location']['lng']
		f['place_id'] = place_id
		f['lat'] = lat
		f['lng'] = lng
		f['description'] = address
		print(f)
		facebook_filtered_with_place_id.append(f)
	except:
		print(f"{address} was not found")
		failed_api_calls.append(f)
		pass
```

```python
# Iterate through the facebook_filtered and update supabase
for f in facebook_filtered_with_place_id:
	# Create the location in the places table using the place_id, lat, lng, description
	supabase.table('places').upsert({'description': f['description'], 'place_id': f['place_id'], 'geog': f"SRID=4326;POINT({lng} {lat})"}).execute()
	# Set the user's place_id in the facebook table to the newly created location
	supabase.table('facebook').update({'email': f['email'], 'place_id': f['place_id']}).eq('email', f['email']).execute()
```
