```python
# Install Supabase
!pip3 install supabase

# Init Supabase and read from .env
import os
from supabase import create_client, Client

url: str = os.environ.get("VITE_SUPABASE_URL")
key: str = os.environ.get("VITE_SUPABASE_ANON_KEY")
google_places_key: str = os.environ.get("VITE_GOOGLE_MAP_KEY_SERVER")
supabase: Client = create_client(url, key)
```

```python
duplicate_places = supabase.table('places').select('*').eq('geog', 'SRID=4326;POINT (-71.10973349999999 42.3736158)').execute().data
```

```python
import pandas as pd

# Write duplicate_places to a pandas dataframe
duplicate_places = pd.DataFrame(duplicate_places)
```

```python
# Save facebook to a pkl
duplicate_places.to_pickle('duplicate_places.pkl')
duplicate_places
```

```python
# Convert to dictionary
duplicate_places_dict = duplicate_places.to_dict('records')
duplicate_places_dict
```

```python
# For each filtered place, get the place ID, description, longitude, and latitude from the Google Places API using the address field
import requests
import json

failed_api_calls = []
for place in duplicate_places_dict[0:]:
	try:
		address = place['description']
		url = f'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={address}&inputtype=textquery&fields=place_id,name,geometry&key={google_places_key}'
		response = requests.get(url)
		data = json.loads(response.text)
		first_result = data['candidates'][0]
		place_id = first_result['place_id']
		lat = first_result['geometry']['location']['lat']
		lng = first_result['geometry']['location']['lng']
		print(first_result)
		print(place_id)
		print(lat)
		print(lng)
		# Create the location in the places table using the place_id, lat, lng, description
		supabase.table('places').upsert({'description': place['description'], 'place_id': place['place_id'], 'geog': f"SRID=4326;POINT({lng} {lat})"}).execute()
	except Exception as e:
		print(e)
		print(f"{address} was not found")
		failed_api_calls.append(place)
		pass
```
