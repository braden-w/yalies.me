```python
# Install Supabase
!pip3 install supabase

# Init Supabase and read from .env
import os
from supabase import create_client, Client

url: str = os.environ.get("VITE_SUPABASE_URL")
key: str = os.environ.get("VITE_SUPABASE_ANON_KEY")
google_places_key: str = os.environ.get("VITE_GOOGLE_MAP_KEY_SERVER")
supabase: Client = create_client(url, key)
```

```python
facebook = supabase.table('facebook').select('*').execute().data
```

```python
import pandas as pd

# Write facebook to a pandas dataframe
facebook = pd.DataFrame(facebook)
```

```python
# Save facebook to a pkl
facebook.to_pickle('facebook.pkl')
facebook
```

```python
# Filter to give the places that have no place ID but a year, address, netid, school, and college
facebook_filtered = facebook[facebook['place_id'].isnull()]
facebook_filtered = facebook_filtered[facebook_filtered['year'].notnull()]
facebook_filtered = facebook_filtered[facebook_filtered['address'].notnull()]
facebook_filtered = facebook_filtered[facebook_filtered['netid'].notnull()]
facebook_filtered = facebook_filtered[facebook_filtered['school'].notnull()]
facebook_filtered = facebook_filtered[facebook_filtered['college'].notnull()]
# Print details about array
print(f"{len(facebook_filtered)} facebook users have no place ID but a year, address, netid, school, and college")
facebook_filtered
```

```python
# Filter this down to just the email and address and place_id columns
facebook_filtered = facebook_filtered[['email', 'address', 'place_id']]
facebook_filtered
```
