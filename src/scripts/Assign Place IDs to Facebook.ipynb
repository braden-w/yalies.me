```python
# Install Supabase
!pip3 install supabase

# Init Supabase and read from .env
import os
from supabase import create_client, Client

url: str = os.environ.get("VITE_SUPABASE_URL")
key: str = os.environ.get("VITE_SUPABASE_ANON_KEY")
google_places_key: str = os.environ.get("VITE_GOOGLE_MAP_KEY_SERVER")
supabase: Client = create_client(url, key)
```

```python
facebook = supabase.table('facebook').select('*').execute().data
facebook
```

```python
# Filter to give the places that have no place ID but a year, address, netid, school, and college
facebook_filtered = [
	    f for f in facebook
	    if f['place_id'] is None and f['year'] is not None and f['address'] is not None and f['netid'] is not None and f['school_code'] is not None and f['school'] is not None and f['college'] is not None
	]
# Print details about array
print(f"{len(facebook_filtered)} facebook users have no place ID but a year, address, netid, school, and college")
facebook_filtered
```

```python

# Filter this down to just the email and address and place_id fields
facebook_filtered = [{ 'email': f['email'], 'address': f['address'], 'place_id': f['place_id'] } for f in facebook]
print(f"{len(facebook_filtered)} facebook users")
len(facebook_filtered)
```

```python
# For each filtered place, get the place ID, description, longitude, and latitude from the Google Places API using the address field
import requests
import json

facebook_filtered_with_place_id = []
for f in facebook_filtered[0:1]:
	address = f['address']
	url = f'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={address}&inputtype=textquery&fields=place_id,name,geometry&key={google_places_key}'
	response = requests.get(url)
	data = json.loads(response.text)
	print(data['candidates'][0])
	place_id = data['candidates'][0]['place_id']
	f['place_id'] = place_id
	facebook_filtered_with_place_id.append(f)
```
