```python
# Install Supabase
!pip3 install supabase

# Init Supabase and read from .env
import os
from supabase import create_client, Client

url: str = os.environ.get("VITE_SUPABASE_URL")
key: str = os.environ.get("VITE_SUPABASE_ANON_KEY")
google_places_key: str = os.environ.get("VITE_GOOGLE_MAP_KEY_SERVER")
supabase: Client = create_client(url, key)
```

```python
facebook = supabase.table('facebook').select('*').execute().data
```

```python
import pandas as pd

# Write facebook to a pandas dataframe
facebook = pd.DataFrame(facebook)
```

```python
# Save facebook to a pkl
facebook.to_pickle('facebook.pkl')
facebook
```

```python
# Filter to give the places that have no place ID but a year, address, netid, school, and college
facebook_filtered = facebook[facebook['place_id'].isnull()]
facebook_filtered = facebook_filtered[facebook_filtered['year'].notnull()]
facebook_filtered = facebook_filtered[facebook_filtered['address'].notnull()]
facebook_filtered = facebook_filtered[facebook_filtered['netid'].notnull()]
facebook_filtered = facebook_filtered[facebook_filtered['school'].notnull()]
facebook_filtered = facebook_filtered[facebook_filtered['college'].notnull()]
# Print details about array
print(f"{len(facebook_filtered)} facebook users have no place ID but a year, address, netid, school, and college")
facebook_filtered
```

```python
# Filter this down to just the email and address and place_id columns
facebook_filtered = facebook_filtered[['email', 'address', 'place_id']]
facebook_filtered
```

```python
# Split the address on the \n character and take the later elements, giving us just the city
facebook_filtered_newline_removed = facebook_filtered.copy()
facebook_filtered_newline_removed['address'] = facebook_filtered_newline_removed['address'].str.split('\n').str[1:].str.join(', ')
facebook_filtered_newline_removed
```

```python
def containsDigit(i: str) -> bool:
	return any(char.isdigit() for char in i)

# Split the address on the ' ' character and remove all elements that contain a number
facebook_filtered_newline_removed_space_removed = facebook_filtered_newline_removed.copy()
facebook_filtered_newline_removed_space_removed['address'] = facebook_filtered_newline_removed_space_removed['address'].str.split(' ')

# Remove all elements that contain a number from the array in address
facebook_filtered_newline_removed_space_removed['address'] = facebook_filtered_newline_removed_space_removed['address'].apply(lambda x: [i for i in x if not containsDigit(i)]).str.join(' ')

# The resulting array has just the city!
facebook_filtered_newline_removed_space_removed
```

```python
# Filter out addresses = ''
facebook_filtered_just_cities = facebook_filtered_newline_removed_space_removed[facebook_filtered_newline_removed_space_removed['address'] != '']
print(f"{len(facebook_filtered_newline_removed_space_removed) - len(facebook_filtered_just_cities)} addresses were removed because they were empty")
```

```python
# Convert to dictionary
facebook_filtered_just_cities = facebook_filtered_just_cities.to_dict('records')
len(facebook_filtered_just_cities)
```

```python
# For each filtered place, get the place ID, description, longitude, and latitude from the Google Places API using the address field
import requests
import json

facebook_filtered_with_place_id = []
failed_api_calls = []
# For f in rows 1 of facebook_filtered_1_new_lines
for f in facebook_filtered_just_cities[0:100]:
	try:
		address = f['address']
		url = f'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?input={address}&inputtype=textquery&fields=place_id,name,geometry&key={google_places_key}'
		response = requests.get(url)
		data = json.loads(response.text)
		place_id = data['candidates'][0]['place_id']
		f['place_id'] = place_id
		facebook_filtered_with_place_id.append(f)
	except:
		print(f"{address} was not found")
		failed_api_calls.append(f)
		pass
```
